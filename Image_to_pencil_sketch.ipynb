```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Reading The Input Image
img = cv2.imread("/content/warrior.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
#Converting Original Image to Grey Scale
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_img)
plt.show()
```

```python
#Inverting the Gray Image
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

```python
#Blurring the inverted gray image by Gausian Function
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

```python
#Inverting the blurred image
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

```python
#Creating Pencil Sketch image
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
plt.axis('off')
plt.imshow(pencil_img)
plt.show()
```

```python
cv2.imwrite("./output_image.png",pencil_img)
```
